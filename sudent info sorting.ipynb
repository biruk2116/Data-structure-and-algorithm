# üîÅ Bubble Sort: Sorts the students by 'id' in ascending order

```python
def bubble_sort_by_id(student_list):
    n = len(student_list)
    for i in range(n):
        for j in range(0, n - i - 1):
            if student_list[j]['id'] > student_list[j + 1]['id']:
                student_list[j], student_list[j + 1] = student_list[j + 1], student_list[j]
    return student_list
```

# ‚úçÔ∏è Insertion Sort: Sorts the students alphabetically by 'name'

```python
def insertion_sort_by_name(student_list):
    for i in range(1, len(student_list)):
        key = student_list[i]
        j = i - 1
        while j >= 0 and key['name'].lower() < student_list[j]['name'].lower():
            student_list[j + 1] = student_list[j]
            j -= 1
        student_list[j + 1] = key
    return student_list
```

# üßÆ Selection Sort: Sorts the students by 'cgpa' in descending order

```python
def selection_sort_by_cgpa(student_list):
    n = len(student_list)
    for i in range(n):
        max_idx = i
        for j in range(i + 1, n):
            if student_list[j]['cgpa'] > student_list[max_idx]['cgpa']:
                max_idx = j
        student_list[i], student_list[max_idx] = student_list[max_idx], student_list[i]
    return student_list
```

# Sorting Algorithms Comparison and Advantage and Disadvantage

## 1. Bubble Sort

### How It Works
- Iterates through the list comparing adjacent elements
- Swaps elements if they are in the wrong order
- Repeats until no more swaps are needed (list is sorted)

### Complexity Analysis
| Case       | Time Complexity | Space Complexity |
|------------|-----------------|------------------|
| Best       | O(n)            | O(1)             |
| Average    | O(n¬≤)           | O(1)             |
| Worst      | O(n¬≤)           | O(1)             |

### Key Characteristics
**Advantages:**
- Simple to understand and implement
- Stable (preserves original order of equal elements)
- Efficient for nearly sorted data (with optimization)
- Works well with small datasets

**Disadvantages:**
- Poor performance on large datasets
- Makes many unnecessary comparisons
- Not cache-efficient due to frequent swaps

### Ideal Use Cases
- Sorting small collections (<100 items)
- When maintaining element order is important
- Situations where data is already mostly sorted

---

## 2. Insertion Sort

### How It Works
- Builds the sorted list one element at a time
- Takes each new element and inserts it into its proper position
- Shifts elements as needed to make space

### Complexity Analysis
| Case       | Time Complexity | Space Complexity |
|------------|-----------------|------------------|
| Best       | O(n)            | O(1)             |
| Average    | O(n¬≤)           | O(1)             |
| Worst      | O(n¬≤)           | O(1)             |

### Key Characteristics
**Advantages:**
- Excellent for small to medium datasets
- Maintains stability of equal elements
- Highly efficient with nearly sorted data
- Can sort data as it arrives (online sorting)
- Makes fewer moves than Bubble Sort

**Disadvantages:**
- Still has quadratic worst-case complexity
- Requires many shifts for unsorted data
- Not the most cache-friendly approach

### Ideal Use Cases
- Sorting small dictionaries or objects
- Maintaining sorted collections with new additions
- Situations where most data is already ordered

---

## 3. Selection Sort

### How It Works
- Divides the list into sorted and unsorted portions
- Repeatedly finds the minimum element in unsorted portion
- Swaps it with the first unsorted element

### Complexity Analysis
| Case       | Time Complexity | Space Complexity |
|------------|-----------------|------------------|
| Best       | O(n¬≤)           | O(1)             |
| Average    | O(n¬≤)           | O(1)             |
| Worst      | O(n¬≤)           | O(1)             |

### Key Characteristics
**Advantages:**
- Minimizes number of swaps (only O(n) swaps)
- Simple conceptual implementation
- Performs well when writes are expensive
- In-place sorting with minimal memory

**Disadvantages:**
- Not a stable sorting algorithm
- Always requires O(n¬≤) comparisons
- Poor performance on large datasets
- Less efficient than Insertion Sort in practice

### Ideal Use Cases
- Situations where write operations are costly
- Small datasets where simplicity is valued
- Cases where stability isn't required

---

## Comparative Analysis

| Feature          | Bubble Sort       | Insertion Sort    | Selection Sort    |
|------------------|-------------------|-------------------|-------------------|
| **Time (Best)**  | O(n)              | O(n)              | O(n¬≤)             |
| **Time (Avg)**   | O(n¬≤)             | O(n¬≤)             | O(n¬≤)             |
| **Stability**    | Stable            | Stable            | Not Stable        |
| **Adaptability** | Good              | Excellent         | None              |
| **Swap Count**   | High              | Medium            | Low               |
| **Best For**     | Nearly-sorted     | Small datasets    | Write-sensitive   |

---

## Practical Recommendations

### For Dictionary Sorting:
1. **First Choice:** Insertion Sort  
   - Best balance of stability and performance
   - Particularly efficient for small datasets
   - Maintains element order relationships

2. **Special Cases:**
   - Bubble Sort (optimized) when data is >90% sorted
   - Selection Sort when minimizing writes is critical

3. **Large Datasets:**
   - Use Python's built-in `sorted()` (Timsort)
   - Consider Merge Sort or QuickSort alternatives

### General Guidance:
- Insertion Sort is typically the best performer among these three for real-world small datasets
- Bubble Sort's main advantage is simplicity for teaching
- Selection Sort's swap optimization is rarely worth its drawbacks
